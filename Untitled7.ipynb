```python
x = df_cleaned.drop(columns=['Anomaly'])
y = df_cleaned['Anomaly']
```

```python
x.shape
```

```python
# Assuming you have already separated X and y
# X contains the features, and y contains the target variable (Anomaly)

# Split the data into training and testing sets (e.g., 80% training, 20% testing)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)

# Display the shapes of the resulting splits
print("X_train shape:", x_train.shape)
print("X_test shape:", x_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

# Assume X_train is the training data and X_test is the test data
scaler = StandardScaler()

# Fit the scaler on the training data and transform it
x_train_scaled = scaler.fit_transform(x_train)

# Now, the test data (in real-time scenario, it's unseen data) will be transformed using the scaler fit on the training data
x_test_scaled = scaler.transform(x_test)  # No fitting here, just transform
```

```python
from keras_tuner import HyperModel  # Correct import
from keras_tuner.tuners import RandomSearch
from keras import regularizers 

# Define the model building function for Keras Tuner
class ANNHyperModel(HyperModel):
    def build(self, hp):
        model = Sequential()
        
        # Number of dense layers
        for i in range(hp.Int('num_layers', 1, 4)):  # 1 to 4 layers
            model.add(Dense(
                units=hp.Int('units_' + str(i), min_value=32, max_value=512, step=32),  # Units in layers
                activation=hp.Choice('activation', values=['relu', 'tanh', 'sigmoid']),  # Activation functions
                kernel_regularizer=regularizers.l2(hp.Float('l2_reg', min_value=0.0, max_value=0.01, step=0.001)),  # Regularization
                input_dim=x_train_scaled.shape[1] if i == 0 else None  # Set input dimension for the first layer
            ))
        
        # Output layer
        model.add(Dense(1, activation='sigmoid'))  # Since it's a binary classification
        
        # Optimizer selection
        optimizer = hp.Choice('optimizer', values=['adam', 'sgd', 'rmsprop', 'adagrad'])

        # Compile the model
        model.compile(
            optimizer=optimizer,
            loss='binary_crossentropy',
            metrics=['accuracy']
        )
        
        return model

# Define the tuner
tuner = RandomSearch(
    ANNHyperModel(),
    objective='val_accuracy',  # Optimization objective
    max_trials=10,  # Number of trials
    executions_per_trial=3,  # Number of executions per trial
    directory='my_dir',  # Directory for storing results
    project_name='ANN_tuning'  # Project name
)

# Perform hyperparameter tuning
tuner.search(x_train_scaled, y_train, epochs=10, batch_size=32, validation_data=(x_test_scaled, y_test))

# Get the best model
best_model = tuner.get_best_models(num_models=1)[0]

# Evaluate the best model
best_model.evaluate(x_test_scaled, y_test)
```

```python
# Retrieve the first trial
first_trial_id = list(tuner.oracle.trials.keys())[0]  # Get the ID of the first trial
first_trial = tuner.oracle.get_trial(first_trial_id)  # Get the trial object

# Display the hyperparameters and metrics of the first trial
print("Hyperparameters of the first trial:")
print(first_trial.hyperparameters.values)

print("\nMetrics of the first trial:")
print(first_trial.metrics)
```

```python
# Build a model using the hyperparameters from the first trial
first_trial_hyperparameters = first_trial.hyperparameters.values

def build_model_with_params(hyperparameters):
    model = Sequential()
    
    # Number of layers and their configuration
    for i in range(hyperparameters['num_layers']):
        model.add(Dense(
            units=hyperparameters[f'units_{i}'],
            activation=hyperparameters['activation'],
            kernel_regularizer=regularizers.l2(hyperparameters['l2_reg']),
            input_dim=x_train_scaled.shape[1] if i == 0 else None
        ))
    
    # Add output layer
    model.add(Dense(1, activation='sigmoid'))
    
    # Compile the model
    model.compile(
        optimizer=hyperparameters['optimizer'],
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    
    return model

# Train the model using the first trial's parameters
model = build_model_with_params(first_trial_hyperparameters)
history = model.fit(
    x_train_scaled, y_train,
    epochs=20,
    batch_size=32,
    validation_data=(x_test_scaled, y_test)
)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

# Get predictions on the test set
y_pred_proba = model.predict(x_test_scaled)  # Predicted probabilities
y_pred = (y_pred_proba > 0.5).astype(int)  # Convert probabilities to binary predictions

# Accuracy
accuracy = accuracy_score(y_test, y_pred)

# Precision
precision = precision_score(y_test, y_pred)

# Recall
recall = recall_score(y_test, y_pred)

# F1 Score
f1 = f1_score(y_test, y_pred)

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)

# Classification Report
class_report = classification_report(y_test, y_pred)

# Display the results
print("Evaluation Metrics:")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
print("\nConfusion Matrix:")
print(conf_matrix)
print("\nClassification Report:")
print(class_report)
```
